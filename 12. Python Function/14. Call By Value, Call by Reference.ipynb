```python
# list always pass by reference : 
def test(x):
    x[0]=5
a = [0, 10, 20, 30, 40]
print(a)
test(a)
print(a)
```

```python
# # Pass by Value:
# primitive datatype(int, float, string, )
# tuple, 
    
# # Pass by Reference (Mutable):
# list[] (mutated by its methods), set, dictionary    
```

```python
a = ["penan", "chetan"]
b=a
print(a)
print(b)
print(id(a)==id(b))

b.append("ABC")
print(a)
print(b)
print(id(a)==id(b))
print("List is always pass by reference")
```

```python
# make list : pass by value
a = ["penan", "chetan"]
b=a[:]
print(a)
print(b)
print(id(a)==id(b))

b.append("ABC")
print(a)
print(b)
print(id(a)==id(b))
print("List is always pass by value")
```

```python
a = ["PENAN", "CHETAN"]
b=a
print(id(a)==id(b))

b.append("ABC")
print(id(a)==id(b))

b = ["ABC", "DEF"]
print(id(a)==id(b))
```

```python
## tuple
# set always pass by reference : 
def test(x):
    x.add(5)
a = {0, 10, 20, 30, 40}
print(a)
test(a)
print(a)
```

```python
## dictionary
# dictionary always pass by reference
def test(x):
    x["PQR"]= 50
a={"ABC":5, "DEF":10, "EFG":15}
print(a)
test(a)
print(a)
```
